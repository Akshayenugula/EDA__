### Done WebScrapping by using the librarys Beautifulsoup and Requests and have scrapped the information of the cars. By using the pandas saved it in to a csv file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("carwale_used_cars.csv")
pd.set_option('display.max_rows', None)
```

```python
# Extract year, brand, and model
df['year'] = df['title'].str.extract(r'(\d{4})')
df['brand'] = df['title'].str.extract(r'^\d{4} (\w+)')
df['model'] = df['title'].str.replace(r'^\d{4} \w+ ', '')
```

```python
df.head()
```

```python
df = df.drop(columns = ['title'])
```

```python
df = df.drop(columns = ['emi'])
```

```python
df.head()
```

```python
df = df.iloc[:, ::-1]
```

```python
df.head()
```

```python
df['location'] = df['location'].str.replace(',','').str.replace('Hyderabad','')
```

```python
df['mileage'] = df['mileage'].str.replace('km','').str.replace(',','')
```

```python
df.head()
```

```python
df = df.rename(columns={'mileage':'KM Driven'})
```

```python
df['price'] = df['price'].str.replace('Rs.', '')
```

```python
def convert_price(price):

    price = price.replace(',', '')
    if 'Lakh' in price:
        return float(price.replace('Lakh', '')) * 100000
    elif 'Crore' in price:
        return float(price.replace('Crore', '')) * 10000000
```

```python
df['price in RS'] = df['price'].apply(convert_price)
df.dropna(subset=['price in RS'], inplace=True)
df['price in RS'] = df['price in RS'].astype(int)
```

```python
df['price in RS'] = df['price'].apply(convert_price).astype(int)
```

```python
df.head()
```

```python
df= df.drop(columns =['price'])
```

# Univariate Analysis

### What is the distribution of car prices? Are most cars affordable or expensive?

```python
plt.figure(figsize=(10,5))
sns.histplot(df['price in RS'], bins=20, kde=True)
plt.title('Distribution of Car Prices')
plt.show()
```

```python
# Handling Outliers: Removing extreme price values
df = df[df['price in RS'] < df['price in RS'].quantile(0.99)]  # Remove top 1% high-priced cars

# Univariate Analysis: Updated price distribution
plt.figure(figsize=(10,5))
sns.histplot(df['price in RS'], bins=20, kde=True)
plt.title('Distribution of Car Prices (After Outlier Removal)')
plt.xlabel('Price in Lakh')
plt.ylabel('Count')
plt.show()

# Log Transformation for better visualization
df['log_price'] = np.log1p(df['price in RS'])  # log1p to handle zero values safely
plt.figure(figsize=(10,5))
sns.histplot(df['log_price'], bins=20, kde=True)
plt.title('Log-Transformed Car Prices Distribution')
plt.xlabel('Log Price')
plt.ylabel('Count')
plt.show()
```

Hereâ€™s a refined version of your summary with better clarity and explanation:  

---

I analyzed the distribution of car prices and found that most cars fall within an **affordable price range**, but there are some **extreme high values (outliers)** in the dataset. To handle these **outliers**, I removed the **top 1% of the most expensive cars**, ensuring a more representative distribution.  

### **Why is transformation needed?**  
The price data was **highly right-skewed**, meaning a small number of very expensive cars distorted the overall distribution. Applying a **log transformation** helped by:  
1. **Reducing skewness**, making the distribution closer to normal.  
2. **Improving visualization**, so price differences across most cars are clearer.  
3. **Enhancing interpretability**, making trends easier to analyze.  

### **Post-transformation insights:**  
After applying the **log transformation**, most cars are priced between **12 to 16 lakh**, with the highest concentration between **13 to 15 lakh**. The distribution remains **slightly right-skewed**, but it is now **more balanced and easier to analyze**.

## What are the correlations between numerical variables like year, KM Driven, and price?

```python
# Selecting numerical columns
numeric_cols = ['year', 'KM Driven', 'price in RS']

# Compute correlation matrix
correlation_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# Selecting numerical columns
numeric_cols = ['year', 'KM Driven', 'price in RS', 'log_price']

# Compute correlation matrix
correlation_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, square=True)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```


The heatmap represents the **correlation** between numerical variables in the dataset. The values range from **-1 to 1**, where:  
- **1.00** â†’ Perfect positive correlation (as one variable increases, the other also increases).  
- **-1.00** â†’ Perfect negative correlation (as one variable increases, the other decreases).  
- **0.00** â†’ No correlation.  

#### **Key Observations from the Image:**  

1. **Year vs. Price (0.31)**  
   - A **positive correlation (0.31)** suggests that **newer cars** tend to be **more expensive**, though not strongly.  
   - This makes sense because newer models often have **better features, lower mileage, and higher demand**.  

2. **Year vs. KM Driven (-0.46)**  
   - A **moderate negative correlation (-0.46)** means **older cars** generally have **higher mileage**.  
   - This is expected, as cars that have been used for a longer time typically have **more kilometers driven**.  

3. **KM Driven vs. Price (-0.35 to -0.40)**  
   - A **negative correlation** indicates that **higher mileage cars tend to have lower prices**.  
   - This suggests that buyers **prefer cars with lower mileage**, as they are perceived as **less used and in better condition**.  

4. **Price vs. Log Price (0.99+)**  
   - A strong correlation exists because **log_price** is just the **log-transformed price**, which helps normalize the data for better analysis.  

### **Conclusion:**  
- **Newer cars are generally more expensive.**  
- **Higher mileage cars tend to have lower prices.**
